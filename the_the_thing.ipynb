```python
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")

if sorted(s1) == sorted(s2):
    print("The strings are balanced (all characters are the same)")
else:
    for i in s1:
        if i in s2:
            print("They are not complementary")
            break
    else:
        print("They are complementary")
```

```python
def pair(l,t):
    L= []
    for i in l:
        if t-i in l:
            L.append((i,t-i))
            l.remove(t-i)
    return(L)

l = eval(input("Enter list of numbers\ne.g. format: [1,2,3,4]"))
t = int(input("Enter target number: "))
print(pair(l,t))
```
